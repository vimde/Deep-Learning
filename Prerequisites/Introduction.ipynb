```python
files = [os.path.join('img_align_celebrity', file_i) for file_i in os.listdir('img_align_celebrity') if '.jpg' in file_i]
```

```python
def get_celebrity_images():
    return [plt.imread(file) for file in files]
```

```python
images = get_celebrity_images()
```

```python
plt.imshow(images[0])
```

```python
plt.imshow(images[9])
```

#### A new dimension called 'Batch' dimension will be added to the NumPy array. 
#### It is just the size of the array or the number of images in this case.

```python
data = np.array(images)
```

```python
data.shape
```

```python
mean_image = np.mean(data, axis = 0)
plt.imshow(mean_image.astype(np.uint8))
```

```python
standard_deviation_image = np.std(data, axis = 0)
plt.imshow(standard_deviation_image.astype(np.uint8))
```
